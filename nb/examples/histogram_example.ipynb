#Â Histogram of numbers occuring in a file

An example

```julia
using NetCDF, PyPlot, StatsBase                                         # Load packages
ncfile = NetCDF.open("/data1/ens/ensemble.t.member1.steps1-90.ll.nc")   # Load ncfile
temp = ncfile.vars["t"][:,:,:,1]                                        # Retrieve only first time step

temp = temp[temp .> 0]                                                  # positive values only due to log-histogram
bins = exp.(range(log(minimum(temp)),log(maximum(temp)),length=1000))   # define 1000 log-spaced bins
H = fit(Histogram,temp,bins)                                            # calculate a histogram

plot(bins[1:end-1],H.weights)                                           # plot bins vs weights
xlabel("Temperature [K]")                                               # label x axis
ylabel("N");                                                            # label y axis
```
